IMPORTAR LOS DATOS:

```python
import pandas as pd

categories = pd.read_csv("../input/categories.csv", header = 0, index_col=False)
customers = pd.read_csv("../input/customers.csv", header = 0, index_col=False)
employees = pd.read_csv("../input/employees.csv", header = 0, index_col=False)
orders = pd.read_csv("../input/orders.csv", header = 0, index_col=False)
order_details = pd.read_csv("../input/order_details.csv", header = 0, index_col=False)
products = pd.read_csv("../input/products.csv", header = 0, index_col=False)
suppliers = pd.read_csv("../input/suppliers.csv", header = 0, index_col=False)

print("Categories:")
display(categories.head())

print("Customers:")
display(customers.head())

print("Employees:")
display(employees.head())

print("Orders:")
display(orders.head())

print("Order details:")
display(order_details.head())

print("Products:")
display(products.head())

print("Suppliers:")
display(suppliers.head())
```

Ventas
orderid,fecha,empleado,producto,idproducto,cantidad,preccioventa,descuento,importe,categoria,proveedor,ciudad_cliente,pais_cliente,cliente

```python

result = (
    orders.merge(employees, on='EmployeeID', how='left')
    .merge(order_details, on='OrderID', how='inner')
    .merge(products, on='ProductID', how='left')
    .merge(categories, on='CategoryID', how='left')
    .merge(suppliers, on='SupplierID', how='left')
    .merge(customers, on='CustomerID', how='left')
)

result['importe'] = ((result['UnitPrice_x'] - (result['Discount'] * result['UnitPrice_x'])) * result['Quantity']).round(2)

final_result = result[[
    'OrderID',
    'OrderDate',
    'FirstName',
    'LastName',
    'ProductID',
    'ProductName',
    'Quantity',
    'UnitPrice_x',
    'Discount',
    'importe',
    'CategoryName',
    'CompanyName_x',
    'City',
    'Country',
    'ContactName_y'
]]

display(final_result)
```

VENTAS TIEMPOS 
cliente,fecha_venta,fecha_venta_anterior,diferencia_dias,diferencia_meses,monto_venta

```python
orders['OrderDate'] = pd.to_datetime(orders['OrderDate'])

result = (
    orders
    .merge(customers, on='CustomerID', how='left')
    .merge(order_details, on='OrderID', how='inner')
    .merge(products, on='ProductID', how='left')
)

# Obtener la compra anterior
def get_previous_order_date(row, orders_df):
    previous_orders = orders_df[
        (orders_df['CustomerID'] == row['CustomerID']) & (orders_df['OrderDate'] < row['OrderDate'])
    ]
    if not previous_orders.empty:
        return previous_orders['OrderDate'].max()  
    return pd.NaT


result['OrderDate_anterior'] = result.apply(lambda row: get_previous_order_date(row, orders), axis=1)

# Diferencias de dias y meses
result['diferencia_dias'] = (result['OrderDate'] - result['OrderDate_anterior']).dt.days
result['diferencia_meses'] = ((result['diferencia_dias'] // 30).fillna(0)).astype(int)

result['monto_venta'] = ((result['UnitPrice_x'] - (result['Discount'] * result['UnitPrice_x'])) * result['Quantity']).fillna(0)

# agrupar los detalles
final_result = (
    result
    .groupby(
        ['OrderID', 'ContactName', 'OrderDate', 'OrderDate_anterior'],
        as_index=False
    )
    .agg(monto_venta=('monto_venta', 'sum'),
         diferencia_dias=('diferencia_dias', 'first'),
         diferencia_meses=('diferencia_meses', 'first'))
)

final_result.rename(columns={
    'OrderDate': 'fecha_venta',
    'OrderDate_anterior': 'fecha_venta_anterior',
    'ContactName': 'cliente'
}, inplace=True)

display(final_result)
```
